On donne la première normalisation (MinMax), pour aider la prise en main des fonctionnalités pandas:

```python
import copy
import pandas as pd
import numpy as np

def normaMinMax(X):
    Xn = X.copy() # on duplique pour éviter de modifier le paramètre
    Xn = Xn.drop(columns="Fruit") # pour ne pas tenir compte de cette information
    nbl, nbc = Xn.shape      # les dimensions du tableau de donnees
    for i in range(nbl):     # on normalise chaque exemple, ligne
        for j in range(nbc): # on normalise chaque variable de l'exemple, colonne
            Xn.iloc[i,j] = ( X.iloc[i,j] - X.iloc[:,j].min() ) / (X.iloc[:,j].max() - X.iloc[:,j].min())
    Xn["Fruit"] = X["Fruit"]
    return Xn

#creation de la dataFrame 
df = pd.DataFrame(dataList[1:21], columns=['Weight', 'Lenght', 'Fruit'])
df = df.astype(int)
print(df)
print(normaMinMax(df))
```

Il est conseillé d'écrire ici deux fonctions pour répondre à ces questions, chacune prenant en paramètre la matrice des données, car elles pourront être réutilisées par la suite pour vérifier que les valeurs d'une matrice ont été complétées.

```python
from sklearn.datasets import load_digits
import random
import math


digits = load_digits()
miss_digits = load_digits()
nbmissingvalues = random.randint(1000, 2000)
for _ in range(nbmissingvalues):
    lig = random.randint(0, 1763)
    col = random.randint(0, 63)
    miss_digits.data[lig, col] = math.nan



def count_total_missing_values(data):
    return np.sum(np.isnan(data))

def count_missing_values_per_column(data):
    return np.sum(np.isnan(data), axis=0)

# Calculer le nombre total de valeurs manquantes
total_missing = count_total_missing_values(miss_digits.data)
print(f"Nombre total de valeurs manquantes : {total_missing}")

# Calculer le nombre de valeurs manquantes par colonne
missing_per_column = count_missing_values_per_column(miss_digits.data)
print("Nombre de valeurs manquantes par colonne :")
print(missing_per_column)

total_missing_calculated = np.sum(count_missing_values_per_column(miss_digits.data))
print(f"Nombre total de valeurs manquantes calculé à partir des colonnes : {total_missing_calculated}")
```

On veut maintenant comparer trois méthodes pour compléter ces valeurs manquantes:

1. Complétion stationnaire (valeur la plus fréquente pour cette colonne)
2. Complétion par combinaison linéaire (moyenne des valeurs de cette colonne pour la classe de l'exemple concerné)
3. Complétion par l'algorithme des $k$ plus proches voisins

Pour mesurer la qualité de chaque méthode, nous calculons le score d'écart moyen au carré $e_m$ pour chaque méthode $m$. Pour chaque ligne $i$ et colonne $j$, soit $\acute{x_{i,j}}$ la valeur imputée par $m$  à la variable $x_{i,j}$ qui était manquante. Alors nous avons $$e_m = \sum_{i,j} (x_{i,j}-\acute{x_{i,j}})^2$$
Plus $e_m$ est petit, meilleure est la méthode $m$: s'assurer de bien comprendre cela (sinon le prof est là...)

(Notons que dans le cas de cet exercice, nous avons dans `digits` les vraies valeurs qui manquent dans `miss_digits`, mais en général nous ne pouvons pas puisque les valeurs manquent. Ici, l'objectif est de comparer la qualité des méthodes dans ce cas de données. Nous verrons dans d'autres TPs comment choisir la meilleure méthode dans le cas général).

L'objectif de cet exercice est de réaliser ces trois méthodes, soit en les codant, soit en récupérant ce qui existe dans `sklearn` ou dans `pandas` (mais alors il faut être capable de lire la documentation). 

Indiquer le(s) programme(s) permettant de compléter les valeurs manquantes selon les trois méthodes ci-dessus, et les ordonner de la meilleure à la moins bonne. 

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
import random
from sklearn.impute import KNNImputer

# Charger les données digits
digits = load_digits()
data_original = digits.data  # Les données originales sans valeurs manquantes
data_original = data_original.astype(float)

# Ajouter des valeurs manquantes dans les données d'origine pour évaluer la qualité de l'imputation
def introduce_missing_values(data, num_missing):
    data_with_missing = data.copy()
    num_samples, num_features = data.shape
    missing_indices = []
    for _ in range(num_missing):
        sample_idx = random.randint(0, num_samples - 1)
        feature_idx = random.randint(0, num_features - 1)
        data_with_missing[sample_idx, feature_idx] = np.nan
        missing_indices.append((sample_idx, feature_idx))  # Conserver les indices des valeurs manquantes
    return data_with_missing, missing_indices

# Générer des données avec des valeurs manquantes
num_missing = random.randint(1000, 2000)
data_with_missing, missing_indices = introduce_missing_values(data_original, num_missing)

# Convertir les données en DataFrame Pandas pour faciliter les manipulations
mdd1 = pd.DataFrame(data_with_missing.copy())  # pour tester méthode 1
mdd2 = pd.DataFrame(data_with_missing.copy())  # pour tester méthode 2
mdd3 = pd.DataFrame(data_with_missing.copy())  # pour tester méthode 3

# Fonction pour calculer le score d'écart moyen au carré em
def calculate_em_score(original_data, imputed_data, missing_indices):
    em_score = 0
    for (i, j) in missing_indices:
        em_score += (original_data[i, j] - imputed_data[i, j]) ** 2
    return em_score

# Définir les fonctions d'imputation
def fill_missing_with_mode(df):
    """Compléter les valeurs manquantes avec le mode (valeur la plus fréquente) de chaque colonne."""
    mode_values = df.mode().iloc[0]  # Assurez-vous que df est un DataFrame Pandas
    return df.fillna(mode_values)

def fill_missing_with_class_mean(df, target):
    """Compléter les valeurs manquantes par la moyenne des valeurs pour chaque classe."""
    for col in df.columns:
        for cls in target.unique():
            mean_value = df.loc[target == cls, col].mean()
            df.loc[(target == cls) & (df[col].isna()), col] = mean_value
    return df

# Initialiser l'imputeur KNN 
knn_imputer = KNNImputer(n_neighbors=5)
def impute_with_knn(df):
    """Compléter les valeurs manquantes en utilisant l'algorithme des k plus proches voisins."""
    return pd.DataFrame(knn_imputer.fit_transform(df), columns=df.columns)

# Appliquer et évaluer chaque méthode
print("\nÉvaluation des méthodes d'imputation (score d'écart moyen au carré) :")

# Imputation par la valeur la plus fréquente (mode) : le moins performant
mdd1_filled = fill_missing_with_mode(mdd1)
em_mode = calculate_em_score(data_original, mdd1_filled.to_numpy(), missing_indices)
print(f"Score d'écart moyen au carré pour l'imputation par mode : {em_mode}")

# Imputation par la moyenne des valeurs de la classe : le deuxième plus performant
target = pd.Series(digits.target)
mdd2_filled = fill_missing_with_class_mean(mdd2, target)
em_class_mean = calculate_em_score(data_original, mdd2_filled.to_numpy(), missing_indices)
print(f"Score d'écart moyen au carré pour l'imputation par la moyenne des valeurs de la classe : {em_class_mean}")

# Imputation par KNN : le plus performant
mdd3_filled = impute_with_knn(mdd3)
em_knn = calculate_em_score(data_original, mdd3_filled.to_numpy(), missing_indices)
print(f"Score d'écart moyen au carré pour l'imputation par KNN : {em_knn}")

# Déterminer la méthode la plus efficace
errors = {
    "Mode": em_mode,
    "Class Mean": em_class_mean,
    "KNN": em_knn
}

best_method = min(errors, key=errors.get)
print(f"\nLa méthode la plus efficace est : {best_method} avec un score d'écart moyen au carré de {errors[best_method]}")
```
